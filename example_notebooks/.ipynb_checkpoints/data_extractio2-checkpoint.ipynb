```python
# создадим пустой датафрейм с колонками
mydf = pd.DataFrame(columns=list(df))
print(mydf.shape)
mydf
```

```python
from tqdm import tqdm
# удалим пустые строки по колонке: 'is_ethicity_superior_meaning'
df1 = df[pd.notnull(df['is_ethicity_superior_meaning'])]
# определяем распределение лейблов 
#ttt=df1['is_ethicity_superior_meaning'].unique()
    
# сделаем цикл по всем этничностям
for my_etn in tqdm(etn_list):
    # производим фильтрацию по конкретной этничности? то есть получаем разне датасеты с одной этничностью
    df_filtr = df1[df1['eth_group_to_code']==my_etn]
    # теперь определим список уникальных документов для конкретной этничности
    doc_list = df_filtr['document.id'].unique()
    #doc_list.astype('int') 
    #print(doc_list)
    #print()
    #doc_list1=[883,571281,1217124]
    # теперь организуем цикл по уникальным документам
    for cur_doc in doc_list:
        # фильтрируем по конкретному  id, то есть получаем разные датасеты с разными id-ками
        # но у каждого даатсета, одна этничность
        doc_filtr = df_filtr[df_filtr['document.id']==cur_doc]
        # список уникальных лейблов для выбранной коллекции документов
        unic_labels = doc_filtr['is_ethicity_superior_meaning'].unique()
        #print()
        # определяем количество лейблов и из количество
        vc = doc_filtr['is_ethicity_superior_meaning'].value_counts()
        #vc = vc.sort_index()
        #print(vc, cur_doc)
        # определяем максимальную величину по количеству лейблов
        mymax = doc_filtr['is_ethicity_superior_meaning'].value_counts().max()
        # определяем индекс максимальной величины
        max_index = vc.idxmax()
        #print('index max value: ', max_index, ' max value :', mymax)
        mymedian = vc.median()
        #print('median: ', mymedian)
        #print()
        # проверка на то что количество лейблов больше 1
        if mymax>1:
             # проверка на max>median
            if mymax>mymedian:
                #print('max>median')
                # выделяем даатсет с заданным лейблом - max_index - значение лейбла с максимальным количеством
                cur_lab = doc_filtr[doc_filtr['is_ethicity_superior_meaning']==max_index]
                #  добавляем строку с с лейблом у которого максимум
                mydf.loc[len(mydf.index)] = np.array(cur_lab.iloc[0])
                #print('добавляем строку по максимуму')
            else: 
                #print('max = < median')
                # нужно проверить на количество лейблов
                # если лейблы один то добавляем  строку, если больше одного то не добавляем
                #print('количество лейблов: ', len(vc))
                if len(vc)==1:
                    # выделяем даатсет с заданным лейблом - max_index - значение лейбла с максимальным количеством
                    cur_lab = doc_filtr[doc_filtr['is_ethicity_superior_meaning']==max_index]
                    #  добавляем строку с с лейблом у которого максимум
                    mydf.loc[len(mydf.index)] = np.array(cur_lab.iloc[0])
                    #print('добавляем строку так как один тип лейбла')
                
```
